```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from citipy import citipy
import numpy as np
import openweathermapy.core as ow
```

```python
lats = np.random.uniform(-90, 90, 1400)
lngs = np.random.uniform(-180, 180, 1400)
#lats
#lngs
```

```python
lats_lngs = zip(lats, lngs)
city_list = []
for lat, lng in lats_lngs:
    city = citipy.nearest_city(lat, lng).city_name
    city_list.append(city)
    print(lat, lng)
```

```python
unique_cities = np.unique(city_list, axis=0)
unique_cities = unique_cities.tolist()
unique_cities
```

```python
cities_coord = zip(unique_cities, lats, lngs)
cities_coord = list(cities_coord)
cities_coord
```

```python
cities_coord_df = pd.DataFrame(cities_coord, columns=['city', 'lat', 'lng'])
cities_coord_df.head(10)
```

```python
api_key = "25bc90a1196e6f153eece0bc0b0fc9eb"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = url + "appid=" + api_key + "&units=" + units + "&q="
```

```python
weather_data = []
cities = cities_coord_df['city']

# Loop through the list of cities and perform a request for data on each
for city in cities:
    
    try:
        data = requests.get(query_url + city).json()
        city = data['name']
        cloud_data = data['clouds']['all']
        country = data['sys']['country']
        date = data['dt']
        humidity =data['main']['humidity']
        lat_data = data['coord']['lat']
        lng_data = data['coord']['lon']
        max_temp_data = data['main']["temp_max"]
        wind_speed_data = data['wind']['speed']
        weather_info = {'City': city, 'Cloudiness': cloud_data, 'Country': country, 'Date': date,\
               'Humidity': humidity, 'Latitude': lat_data, 'Longitude': lng_data,
                'Max Temp': max_temp_data, 'Wind Speed': wind_speed_data}
        weather_data.append(weather_info)
    except Exception as e: 
       print(f"{city} was skipped")
       continue
final_weather = pd.DataFrame(weather_data)
```

```python
final_weather
```

```python
plt.scatter(final_weather['Latitude'], final_weather['Max Temp'], marker="o")
plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Temperature (Fahrenheit)")
plt.xlabel("Latitude")
plt.axvline(0, color='purple')
plt.axhline(0, color='green')
plt.grid(True)

# Save the figure
plt.savefig("City_Latitude_vs_Max_Temperature_1.png")

# Show plot
plt.show()
```
